```python
import matplotlib.pyplot as plt
import numpy as np

def funcao(t):
    return 10**(1-t)

n = np.linspace(2,12, num=1000)
plt.plot(n, funcao(n))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def relativo(t):
    return 0.5*((10**(1-t))+1)

def absoluto(t):
    return 0.5*(10**(1-t))

n = np.linspace(2,12, num=1000)
plt.plot(n, relativo(n))
plt.plot(n, absoluto(n))
plt.show()
```
